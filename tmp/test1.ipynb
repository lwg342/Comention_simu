```python
def f(x):
    return np.argsort(x)#[::-1]
```

```python
import random
x = np.array([i for i in range(0, 500)])
random.shuffle(x)
print(x)
# y = range(1, 9)
# Ordery = np.argsort(y)[::-1]
Orderx = f(x)
print(Orderx)
print(f(Orderx))

all( x == f(f(f(f(x)))) ), all( x == f(f(f(x))) ), all( x == f(f(x)) )
```
